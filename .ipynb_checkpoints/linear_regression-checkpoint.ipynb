# Linear Regression
## Global Happiness Rankings (2015, via Kaggle)

### Initial Data Prep

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet

data_df = pd.read_csv('2015.csv')

# Clean the data
data = data_df.dropna()

# Create training/test sets
features = data_df.drop('Happiness Score', axis=1)
labels = data['Happiness Score']
train_set, test_set = train_test_split(data_df, test_size=0.2, random_state=101)
working_set = train_set.copy()
# print(train_set.head())
```

### Initial X for target Y features:
GDP as X and happiness score as Y. I suspect these will be the most closely related.

```python
# Finding the linear regression
reg = ElasticNet(alpha = 0.01, l1_ratio=0.5)
X = working_set[['Economy (GDP per Capita)']]
Y = working_set['Happiness Score']
reg.fit(X, Y)
print("The bias is " , reg.intercept_)
print("The feature coefficients are ", reg.coef_)

# Plotting our results
theta0 = reg.intercept_
theta1 = reg.coef_[0]
print(theta0, theta1)
def the_line(h, bias, coef):
    return bias + coef * h

min_gdp = working_set['Economy (GDP per Capita)'].min()
min_happiness = the_line(min_gdp, theta0, theta1)
max_gdp = working_set['Economy (GDP per Capita)'].max()
max_happiness = the_line(max_gdp, theta0, theta1)
line_x = [min_gdp, max_gdp]
line_y = [min_happiness, max_happiness]

import matplotlib.pyplot as plt
plt.scatter(working_set['Economy (GDP per Capita)'], working_set['Happiness Score'])
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.plot(line_x, line_y)
plt.show()
```

### Analysis:
The linear corellation is there, but there's a lot of variance. Perhaps a different feature will be better.

```python
# Finding the linear regression for a different feature
reg = ElasticNet(alpha = 0.01, l1_ratio=0.5)
X = working_set[['Freedom']]
Y = working_set['Happiness Score']
reg.fit(X, Y)
print("The bias is " , reg.intercept_)
print("The feature coefficients are ", reg.coef_)

# Plotting our results
theta0 = reg.intercept_
theta1 = reg.coef_[0]
print(theta0, theta1)
def the_line(h, bias, coef):
    return bias + coef * h

min_gdp = working_set['Freedom'].min()
min_happiness = the_line(min_gdp, theta0, theta1)
max_gdp = working_set['Freedom'].max()
max_happiness = the_line(max_gdp, theta0, theta1)
line_x = [min_gdp, max_gdp]
line_y = [min_happiness, max_happiness]

import matplotlib.pyplot as plt
plt.scatter(working_set['Freedom'], working_set['Happiness Score'])
plt.xlabel('Freedom')
plt.ylabel('Happiness Score')
plt.plot(line_x, line_y)
plt.show()
```

### Analysis:
Freedom fares even worse. We'll try multiple features.

```python
reg_plus = ElasticNet(alpha = .01, l1_ratio=0.5)

X = working_set[['Economy (GDP per Capita)', 'Freedom']]
Y = working_set['Happiness Score']
reg_plus.fit(X, Y)

print(reg_plus.score(X,Y))
print("The bias is " , reg_plus.intercept_)
print("The feature coefficients are ", reg_plus.coef_)
```

### Analysis:
Slightly better. I think introducing more features would overtrain the model. Let's evaluate it on the test set.

```python
X = test_set[['Economy (GDP per Capita)', 'Freedom']]
Y = test_set['Happiness Score']

print(reg_plus.score(X, Y))
```

### Analysis:
Our model scores a 56.65% currently. This is substandard. I imagine that happiness would have a soft minimum and maximum; maybe we will get better results when we move on to polynomial regression.
