# Initial Exploration
## Global Happiness Rankings (2015, via Kaggle)

### Pulling in data and examining the head/tail:

```python
import pandas as pd
from sklearn.model_selection import train_test_split
data_df = pd.read_csv('2015.csv')
# Initial examinations
print(data_df.keys())
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print(data_df.head())
print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~')
print(data_df.tail())
```

### Interesting relations so far:
* GDP seems to play a huge role in happiness. Region also seems relevant, but my guess is that other factors are affected by region which leads to reduced happiness. I suspect everything will lead back to GDP, which directly affects quality of life.<br>
* I would like to be able to predict happiness of a country based on its GDP per capita and/or freedom rating.

### Cleaning the data, creating training set:

```python
import matplotlib.pyplot as plt

# Clean the data
data = data_df.dropna()

# Create training/test sets
features = data_df.drop('Happiness Score', axis=1)
labels = data['Happiness Score']
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=101)

df = pd.concat([X_train, y_train], axis=1)

# Interesting scatter plots for most features
for col in ['Region', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual']:
    plt.scatter(data[col], data['Happiness Score'], label=col, color='c')

    plt.title(f'Happiness Score vs {col}')
    plt.xlabel(col)
    plt.ylabel('Happiness Score')
    plt.legend()
    if(col == 'Region'):
        plt.xticks(rotation=90)
    plt.show()
```
