# initial_exploration_mk2
Loading up our data:

```python
import pandas as pd
import pickle

with open('df.pkl', 'rb') as f:
    df = pickle.load(f)
```

Viewing some scatter plots to examine relations (we will do in-depth analysis later):

```python
import matplotlib.pyplot as plt

for col in ['Region', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual']:
        plt.scatter(df[col], df['Happiness Score'], label=col, color='c')
        plt.title(f'Happiness Score vs {col}')
        plt.xlabel(col)
        plt.xticks(rotation=90)
        plt.ylabel('Happiness Score')
        plt.show()
        
```

Some of these appear to have some strong linear relations.
